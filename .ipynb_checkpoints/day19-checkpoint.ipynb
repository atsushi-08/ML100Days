```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
"1.   畫出 cos 圖檔，並儲存"
x=np.arange(0,4,0.1)
y=np.cos(x*np.pi)
plt.plot(x, y,color='b')
```

```python
"2.    給定散點圖顏色"
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.scatter(X,Y, c=np.arctan2(X, Y),cmap='rainbow', s=50)
plt.title("Scatter plot ")
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
data = pd.read_csv('/Users/robin/Downloads/boston.csv')
data = pd.DataFrame(data)
data
```

```python
"畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？"
data.boxplot(figsize = (15, 6))
```

```python
median = data.median().between(300, 400)
print(median)
```

```python
"畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係"
x = data['NOX']
y = data['DIS']
plt.scatter(x, y, color = 'b', alpha = 0.3, marker = 'o')
plt.xlabel('NOX')
plt.ylabel('DIS')
plt.title('NOX & DIS')
plt.show()
```
